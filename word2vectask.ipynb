```python
uploaded = files.upload()
```

```python
file_name = list(uploaded.keys())[0]
doc = Document(file_name)
text = "\n".join([para.text for para in doc.paragraphs if para.text.strip() != ""])

print("✅ File loaded successfully!")
print(text[:1000])  # Preview first 1000 characters
```

```python
# STEP 2️⃣ : Preprocessing the text
import re
import nltk
nltk.download('punkt')
nltk.download('punkt_tab')
```

```python
def clean_text(text):
    text = re.sub(r"http\S+", "", text)  # remove links
    text = re.sub(r"[^a-zA-Z\s]", "", text)  # keep letters only
    text = text.lower()  # lowercase
    return text

cleaned_text = clean_text(text)
```

```python
sentences = nltk.sent_tokenize(cleaned_text)
tokenized_sentences = [nltk.word_tokenize(sent) for sent in sentences]

print("✅ Preprocessing done! Total sentences:", len(tokenized_sentences))
print(tokenized_sentences[:3])
```

```python
# STEP 3️⃣ : Train Word2Vec model
!pip install gensim
from gensim.models import Word2Vec
```

```python
model = Word2Vec(
    sentences=tokenized_sentences,
    vector_size=100,  # vector dimensions
    window=5,         # context window size
    min_count=2,      # ignore words appearing <2 times
    workers=4,
    sg=1              # use skip-gram (better for small data)
)

print("✅ Model trained successfully!")
```

```python
model.wv.most_similar("linkedin", topn=10)
```

```python
model.wv.most_similar("client")
```

```python
model.wv.most_similar("money")
```

```python
y = model.wv.index_to_key
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=3)
```

```python
X = pca.fit_transform(model.wv.get_normed_vectors())
```

```python
# STEP 6️⃣ : Save model for later use
model.save("word2vec_linkedin.model")
print("✅ Model saved as word2vec_linkedin.model")
```
